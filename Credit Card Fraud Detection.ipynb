```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression 
from sklearn.metrics import accuracy_score
```

```python
data=pd.read_csv('creditcard.csv')
```

```python
data.head()
```

```python
data.isnull().sum()
```

```python
data.info()
```

```python
data.describe()
```

```python
data['Class'].value_counts()
```

0---> Normal Transactions
1---> Fraud Transactions
Highly imbalanced dataset

```python
legit=data[data.Class ==0]
fraud=data[data.Class ==1]
```

```python
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
data.groupby('Class').mean()
```

```python
# Building a sample dataset Containing similar distribution
```

```python
legit_sample = legit.sample(n=492)
```

```python
# concat two dataframes
```

```python
new_data=pd.concat([legit_sample,fraud],axis=0)
```

```python
new_data.head()
```

```python
new_data['Class'].value_counts()
```

```python
new_data.groupby('Class').mean()
```

```python
#splitting data into features and targets
```

```python
X=new_data.drop(columns='Class',axis=1)
Y=new_data['Class']
```

```python
#split data into training and test data
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2, stratify=Y,random_state=2)
```

## Model Training

```python
#Logestic Regression
```

```python
model=LogisticRegression()
```

```python
model.fit(X_train,Y_train)
```

# Model Evaluation

```python
#Accuracy SCore
```

```python
X_train_pred =model.predict(X_train)
train_data_acc =accuracy_score(X_train_pred,Y_train)
```

```python
print("Accuracy Score on Training data", train_data_acc)
```

```python
X_test_pred=model.predict(X_test)
test_data_acc = accuracy_score(X_test_pred,Y_test)
print("Accuracy Score on Test data",test_data_acc)
```
